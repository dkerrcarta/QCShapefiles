```python
from pathlib import Path
import geopandas as gpd
import numpy as np


BASE_DIR = Path('.').resolve()
hab = BASE_DIR.joinpath('test_data/test_data_in/habitat_shp/test_habitat.shp')
grid = BASE_DIR.joinpath('test_data/test_data_in/tile_shp/test_tiles.shp')

hab_gdf = gpd.read_file(hab)
grid_gdf = gpd.read_file(grid)
print(hab_gdf['Id'][142])
print(grid_gdf['NAME'][4])
```

```python

for index, orthoid in grid_gdf.iterrows():
    tile = gpd.GeoDataFrame({'orthoid': [orthoid['OrthoID']],'geometry': [orthoid['geometry']]})
    gdf_sub = gpd.overlay(tile, hab_gdf, how='intersection')
    gdf_sub.crs = {'init': 'epsg:32639'}
    print(gdf_sub.crs)
```

```python
from shapely.geometry import Point
import geopandas as gpd
import random
import fiona

def random_points_in_polygon(number, polygon):
    points = []
    min_x, min_y, max_x, max_y = polygon.bounds
    i= 0
    while i < number:
        point = Point(random.uniform(min_x, max_x), random.uniform(min_y, max_y))
        if polygon.contains(point):
            points.append(point)
            i += 1
    return points  # returns list of shapely point

for index, orthoid in grid_gdf.iterrows():
    points = random_points_in_polygon(50, orthoid['geometry'])
    gdf = gpd.GeoDataFrame({'ID': range(50), 'geometry': points})
     #'Interpretation':[], 'HabitatNumber': [], 'QC_HabitatClass': [], 'QC_Habitat_Name': [], 'QC_By': [],
    cols = ['ID', 'Interpretation', 'HabitatNumber', 'QC_HabitatClass', 'QC_Habitat_Name', 'QC_By', 'geometry']
    gdf['Interpretation'] = ""
    gdf['HabitatNumber'] = ""
    gdf['QC_HabitatClass'] = ""
    gdf['QC_Habitat_Name'] = ""
    gdf['QC_By'] = ""
    gdf = gdf[cols]
    print(gdf.head())
    
    
```

```python
for i in points:
    assert gdf.contains(i).any()
```
