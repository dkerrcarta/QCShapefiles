```python
import pandas as pd
import geopandas as gpd
import numpy as np
from pathlib import Path

BASE_DIR = Path('.').resolve()
df = pd.read_csv(BASE_DIR.joinpath('make_qc_shapefiles/classes.csv'))
gdf = gpd.read_file(BASE_DIR.joinpath('test/test_data/test_data_out/WV_C18_L12/WV_C18_L12.shp'))
gdf['HabitatSub'] = 9240
gdf
```

```python
df['HabitatSub'] = df['Int_SubNum']
join_df = gdf.merge(df, on='HabitatSub', how='left')
cols_to_keep = gdf.columns
```

```python
join_df['HabitatTyp'] = join_df['Int_num']
join_df['HabitatT_1'] = join_df['Int_cls']
join_df['HabitatS_1'] = join_df['Int_SubCls']
join_df = join_df[cols_to_keep]
join_df
```

```python
join_df['AREA_'] = join_df['geometry'].area / 10**6
join_df['AREA_HA_'] = join_df['geometry'].area / 10**4
```
